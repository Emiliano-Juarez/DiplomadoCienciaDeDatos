Implementación de One Hot Encoding en un data set relacionado a Recursos Humanos, en las variables de "Sexo" y "Ciudadanía".

```python
import pandas as pd
import numpy as np
```

```python
df_rh = pd.read_csv("HRDataset_v14.csv")
```

```python
#Mostrar los datos
df_rh.head(5)
```

```python
# Crear un nuevo data set
df_rh1 = df_rh[["Employee_Name", "Position", "State", "Sex", "DOB", "MaritialDesc",
                "CitizenDesc", "Department". "PerformanceScore", "EngagementSurvey", "EmpSatisfaction"]]
```

```python
# Ver el tamaño
df_rh1.shape
```

```python
# Hacer un describe
df_rh1.describe
```

```python
# Hacer un head()
df_rh1.head()
```

```python
# Crear datos dummys de la columna sex
dummy_sex = pd.get_dummies(df_rh1["Sex"], prefix="sexo")
```

```python
# Vemos que contiene
dummy_sex.head()
```

```python
# Eliminamos la anterior columna
df_rh1 = df_rh1.drop(["Sex"], axis=1)
```

```python
# Vemos el tamaño
df_rh1.shape
```

```python
# Concatenamos el dataframe
df_rh2 = pd.concat([df_rh1, dummy_sex], axis=1)
df_rh2
```

```python
# Vemos el tamaña
df_rh2.shape
```

```python
# Hacemos un head()
df_rh2.head()
```

```python
# Creamos una funcion para cambiar a numerica la ciudadania
def crear_dummies(df, varname):
    dummy = pd.get_dummies(df[varname], prefix=varname)
    df = df.drop(varname, axis=1)
    df = pd.concat([df, dummy], axis=1)
    return df
```

```python
#llamamos la funcion createdummies
#Le pasamos el dataframe que deseamos modificar df_rh2
#Le pasamos la columna que deseamos cambiar "CitizenDesc"
crear_dummies(df_rh2, "CitizenDesc")
df_rh2.head()
```

### Conclusión

En el último resultado podemos observar la implementación completa del ONE HOT ENCODING en las variables de "Sexo" y "Ciudadanía"
