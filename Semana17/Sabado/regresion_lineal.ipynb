# Contexto  y Datos

```python
import pandas as pd
datos = pd.read_csv("ingreso.csv")
datos.head()
```

Variable independiente = horas   
Variable a predecir = ingreso

<BR>
    

```python
import matplotlib.pyplot as plt
plt.ylabel("Ingreso ($)")
plt.xlabel("Promedio de horas semanales trabajadas")
plt.scatter(datos["horas"], datos["ingreso"], color="blue")
plt.show()
```

# Creación del Modelo de
# Regresión Lineal Simple

$$y=mx+b$$

```python
from sklearn import linear_model
regresion = linear_model.LinearRegression() # Con este método nos permitira a ajustar nuestros datos a una regresión lineal

horas = datos["horas"].values.reshape((-1, 1))

# Generando el modelo
modelo = regresion.fit(horas, datos["ingreso"])

print("Intersección (b)", modelo.intercept_) # Estamos obteniendo la intersección
print("Pendiente (m)", modelo.coef_) # Obtener la pendiente
```

```python
# Vamos a calcular el salario para una persona que trabaja las sig horas
entrada = [[39.5], [40], [43], [43.5]]
modelo.predict(entrada)

# Hacer una gráfica para ver nuestras predicciones
plt.scatter(entrada, modelo.predict(entrada), color = "red")
plt.plot(entrada, modelo.predict(entrada), color = "black")
plt.ylabel("Ingresos ($)")
plt.xlabel("Promedio de horas semanales trabajadas")
plt.scatter(datos["horas"], datos["ingreso"], color = "blue", alpha=0.55)
plt.show()
```
